```python
# read questions
questions_file = open("../data/multi_instruct/vqa_v2/v2_OpenEnded_mscoco_train2014_questions.json")
train_questions = json.load(questions_file)
questions = [question_data for question_data in train_questions["questions"] if question_data["question"].startswith("According to the hat, what tourist destination is he visiting?")]
    
questions
```

```python
annotations_file = open("../data/multi_instruct/vqa_v2/v2_mscoco_train2014_annotations.json")
train_annotations = json.load(annotations_file)
annotation = [annotation_data for annotation_data in train_annotations["annotations"] if annotation_data["question_id"] == 171837000]

annotation
```
